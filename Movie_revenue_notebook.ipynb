```python
### Irresepective of other features, Movies aired in late fringe produces greater revenue and early morning prodcues least revenue
```

```python
def fun(row):
    if (row['AIRING_DATE']==5 or row['AIRING_DATE']==6):
        return 'Weekend'
    else:
        return 'Weekday'
    
```

```python
final_df['Weekday_Weekend'] = final_df.apply(lambda row: fun(row), axis=1)
```

```python
def season(row):
    if (row['AIRING_MONTH']==1 or row['AIRING_DATE']==12 or row['AIRING_DATE']==11):
        return 'Winter'
    if(row['AIRING_MONTH']==2 or row['AIRING_MONTH']==3):
        return 'Spring'
    if(row['AIRING_MONTH']>=4 and row['AIRING_MONTH']<=7):
        return 'Summer'
    else:
        return 'Autumn'
        
    
```

```python
final_df['Season'] = final_df.apply(lambda row: season(row), axis=1)
```

```python
sns.boxplot(x='Season',y='AIRING_REVENUE',data=final_df,width=0.5)
```
